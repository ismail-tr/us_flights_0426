```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
flights=sns.load_dataset("flights")    #Load flights dataset
flights["year"]=pd.to_datetime(flights["year"], format="%Y").dt.to_period('Y')                                        # Noticed that the “Year” 
                                                                                                                      # //was not in the correct data
                                                                                                                      # //format.
                                                                                                                      # Transformed it into datetime.
flights_per_year=flights.groupby("year", as_index=False)["passengers"].sum()                                          # Just cheking the data. 
                                                                                                                      # //Not in the pie.
flights["seasons"]=[None]*len(flights["passengers"])                                                                  # Create a new column named 
                                                                                                                      # //seasons. 
winter_months=["Dec", "Jan", "Feb"]                                            
spring_months=["Mar", "Apr", "May"]
summer_months=["Jun", "Jul", "Aug"]
autumn_months=["Sep", "Oct", "Nov"]
flights['seasons'] = np.where(
    flights['month'].isin(winter_months), "Winter",
    np.where(
        flights['month'].isin(spring_months), "Spring",
        np.where(                                                                                                      # Grouped the months by season.
            flights['month'].isin(summer_months), "Summer",
            np.where(
                flights['month'].isin(autumn_months), "Autumn", "Unknown"
            )
        )
    )
)
flights["seasons"]=flights["seasons"].astype("category")                                                                # Set "seasons" column as 
                                                                                                                        # //a category.
flights_by_years_and_seasons=flights.groupby(["year", "seasons"], as_index=False)["passengers"]                         # Again just checked the data.
                                                                                                                        # //Not in the pie.
flights_by_seasons=flights.groupby("seasons", as_index=False)["passengers"].sum()                                       # Grouped
fig, axes = plt.subplots(1, 2, figsize=(14, 6))                                                                         # Necessary to print out 2 
                                                                                                                        # //different graphs.
axes[0].pie(flights_by_seasons["passengers"], labels=flights_by_seasons["seasons"],autopct='%1.0f%%',explode=[0.01,0.01,0.01,0.01],
        colors=['#4A6FA5', '#6B8CBE', '#8CA8D6', '#AEC3E8'])                                                            # First pie.
axes[0].set_title("Passenger numbers by season")
flights_by_months=flights.groupby("month", as_index=False)["passengers"].sum()                                          # Grouped
month_colors=['#F94144', '#F9844A', '#F8961E', '#F9C74F', '#90BE6D', '#43AA8B', 
          '#4D908E', '#577590', '#E9C46A', '#E76F51', '#F4A261', '#2A9D8F']
axes[1].pie(flights_by_months["passengers"], labels=flights_by_months["month"], autopct="%1.0f%%",colors=month_colors)  # Second pie.
axes[1].set_title("Passenger numbers by month")
plt.show()
```
